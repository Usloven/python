Найти интервалы, на которых функция убывает

Определить промежутки, на котором f < 0

````python
from sympy import *
x = Symbol('x', real=True)
expr =  - 18*x**3+5*x**2 + 10*x - 30 
expr1 = -12*x**4*sin(cos(x)) 
expr2 = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30 
````

````python
solve(expr,x, real=True) #корень 
````

````python
plot (expr, (x, -1000, 1000)) #графики функций на разных интервалах и графики разных частей функции
plot (expr1, (x, -2, 2))
plot (expr2, (x, -2, 2))
plot (expr1, -100, 100)
plot (expr2, -100, 100)
````
